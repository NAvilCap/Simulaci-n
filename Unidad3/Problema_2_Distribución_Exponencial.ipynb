```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetro de la distribución exponencial (tasa de carga por coche)
lambda_carga = 1 / 15  # Tiempo promedio de carga = 15 minutos

# Función para generar un tiempo de inactividad
def generar_tiempo_carga(lambda_carga):
    u = np.random.uniform(0, 1)  # Generar número aleatorio uniforme
    # Aplicar transformada inversa
    tiempo_carga = -1 / lambda_carga * np.log(1 - u)
    return tiempo_carga

# Generar 12 horas de carga que es igual a 48 veces
tiempos_carga = [generar_tiempo_carga(lambda_carga)
                       for _ in range(48)]

# Mostrar los tiempos de carga generados
print("Tiempos de carga generados:", tiempos_carga)

# Graficar los tiempos de carga simulados
plt.hist(tiempos_carga, bins=10, edgecolor='black', alpha=0.7)
plt.title('Distribución de Tiempos de carga Simulados (Exponencial)')
plt.xlabel('Tiempo de carga (minutos)')
plt.ylabel('Frecuencia')
plt.show()
```

La mayoria de los tiempos de llegada de los carros para cargar se centran en alrededor de menos de 10 minutos
