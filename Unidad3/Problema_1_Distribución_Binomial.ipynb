```python
import numpy as np
import matplotlib.pyplot as plt
from math import comb

# Parámetros de la distribución binomial
n = 20   # Número de visitantes diarios
p = 0.70 # Probabilidad de que el stock este con los articulos solicitados

# Generar un número aleatorio uniforme U ~ Uniform(0, 1)
num_simulaciones = 7  # Número de simulaciones (días)
U = np.random.uniform(0, 1, num_simulaciones)

# Función de distribución acumulada de la distribución binomial
def binomial_pmf(n, k, p):
    return comb(n, k) * (p**k) * ((1 - p)**(n - k))

def binomial_inverse_transform(n, p, U):
    F_k = 0  # Inicializamos la FDA
    k = 0
    while True:
        F_k += binomial_pmf(n, k, p)  # Sumamos la probabilidad hasta que F(k) >= U
        if U <= F_k:
            return k  # Este valor de k es el número de pedidos
        k += 1

# Generar la cantidad de clientes que realizaron un pedido en cada simulación (día)
compras_simuladas = [binomial_inverse_transform(n, p, u) for u in U]

# Mostrar resultados
print("Pedidos Satisfechos por dia de la semana (día):")
print(compras_simuladas)
print(f"Media teórica de la distribución: {n * p}")  # Mostrar la media teórica
# Graficar los resultados

plt.bar(range(1, num_simulaciones + 1), compras_simuladas, tick_label=[f'Día {i+1}' for i in range(num_simulaciones)])
plt.title("Simulación de número de pedidos diarios (Distribución Binomial)")
plt.xlabel("Dias")
plt.ylabel("Pedidos Satisfechos por Dia")
plt.show()
```

Al poder observar en la grafica anterior, obtenemos la simulacion de pedidos satisfechos por dia. En donde la media de los dias en este ejemplo es de alrededor de 13 pedidos, mientras que la media teorica es de 14, siendo un valor cercano.
Como conclusion podemos intuir que el stock indicado por semana es de 90 pedidos satisfechos.
