```python
# Semilla inicial
semilla = 1507
n = 4
iteraciones = 100
resultados = []

for _ in range(iteraciones):
    # Elevar la semilla al cuadrado
    cuadrado = str(semilla ** 2).zfill(2 * n)  #Agregar los espacios necesarios
    # Extraer los dígitos centrales
    inicio = (len(cuadrado) - n) // 2
    semilla = int(cuadrado[inicio:inicio + n])
    resultados.append(semilla)

# Imprimir los resultados
for i, numero in enumerate(resultados, 1):
    print(f"Número Aleatorio {i}: {numero}")
```

```python
def generar_numeros_pseudoaleatorios(cantidad, semilla, a, c, m):
    numeros = []
    x = semilla  # Valor inicial (semilla)
    for _ in range(cantidad):
        x = (a * x + c) % m  # Fórmula del método congruencial lineal
        numeros.append(x)  # Agregar el número generado a la lista
    return numeros

# Parámetros del método congruencial lineal
cantidad = 100  # Cantidad de números pseudoaleatorios a generar
semilla = 12345  # Valor inicial (puede ser cambiado a tu conveniencia)
a = 1664525  # Multiplicador en la fórmula
c = 1013904223  # Incremento en la fórmula
m = 2**16  # Módulo
# Generar los números
numeros_aleatorios = generar_numeros_pseudoaleatorios(cantidad, semilla, a, c, m)

# Imprimir los números generados
for numero in numeros_aleatorios:
    print(numero)
```
